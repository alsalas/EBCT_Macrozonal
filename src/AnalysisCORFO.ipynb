```python
import pandas as pd
import altair as alt

# 1) Leer el CSV asumiendo tabulaciones como separador
df = pd.read_csv('../run/corfo_bb.csv', sep='\t')

# 2) Limpiar y convertir la columna de financiamiento a entero
df['Financiamiento Innova'] = (
    df['Financiamiento Innova']
    .astype(str)
    .str.replace('[\$\.\,]', '', regex=True)  # elimina $ y separadores de miles
    .fillna('0')
    .astype(int)
)

# 3) Configurar Altair para conjuntos grandes
alt.data_transformers.disable_max_rows()

# 4) Crear gráfico de barras: suma de financiamiento por instrumento
chart = (
    alt.Chart(df)
    .mark_bar(cornerRadiusTopLeft=3, cornerRadiusTopRight=3)
    .encode(
        y=alt.X('Instrumento Homologado:N', sort='-x', title='Instrumento'),
        x=alt.Y('sum(Financiamiento Innova):Q', title='Total Financiamiento (CLP)'),
        color='Año Adjudicación:N',
        tooltip=[
            alt.Tooltip('Instrumento:N', title='Instrumento'),
            alt.Tooltip('sum(Financiamiento Innova):Q', title='Financiamiento')
        ]
    )
    .properties(width=400, height=400)
).facet('Región')

chart.show()
```

```python
import pandas as pd
import altair as alt

# 1) Leer el CSV asumiendo tabulaciones como separador
df = pd.read_csv('../run/corfo_bb.csv', sep='\t')

# 2) Limpiar y convertir la columna de financiamiento a entero
df['Financiamiento Innova'] = (
    df['Financiamiento Innova']
    .astype(str)
    .str.replace('[\$\.\,]', '', regex=True)  # elimina $ y separadores de miles
    .fillna('0')
    .astype(int)
)

# 3) Configurar Altair para conjuntos grandes
alt.data_transformers.disable_max_rows()

# 4) Crear gráfico de barras: suma de financiamiento por instrumento
chart = (
    alt.Chart(df)
    .mark_bar(cornerRadiusTopLeft=3, cornerRadiusTopRight=3)
    .encode(
        y=alt.X('Sector Económico:N', sort='-x', title='Instrumento'),
        x=alt.Y('sum(Financiamiento Innova):Q', title='Total Financiamiento (CLP)'),
        color='Año Adjudicación:N',
        tooltip=[
            alt.Tooltip('Instrumento:N', title='Instrumento'),
            alt.Tooltip('sum(Financiamiento Innova):Q', title='Financiamiento')
        ]
    )
    .properties(width=400, height=600)
).facet('Región')

chart.show()
```

```python
from difflib import SequenceMatcher
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Sólo se debe cambiar este número entre 0-6 para seleccionar las áreas
# 3, 10, 41, 13, 4, 2, 24, 22
choose = 22
areas_oecd = df['Sector Económico'].unique()
for i,area in enumerate(areas_oecd):
    print(i,": ", area)

# 1) Filtrar el DataFrame correctamente
df2 = df[
    (df['Región'] == 'Región del Biobío') &
    (df['Sector Económico'] == areas_oecd[choose])
].copy()
num_projects = len(df2)

# 2) Preparar la lista de palabras a partir de la columna 'Tendencia Final'
text_series = df2['Tendencia Final'].astype(str)
words = [
    palabra
    for texto in text_series
    for palabra in texto.split()
    if palabra != 'tendencia'
]
words = [x.replace('(', '') for x in words]
words = [x.replace(')', '') for x in words]

# 3) Construir la matriz de similitud
n = len(words)
sim_matrix = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        sim_matrix[i, j] = (
            1.0 if i == j
            else SequenceMatcher(None, words[i], words[j]).ratio()
        )

# 4) Calcular el índice de similitud medio por palabra
mean_sim = sim_matrix.mean(axis=1)

# 5) Generar el diccionario palabra → peso
freq_dict = { words[i]: mean_sim[i] for i in range(n) }

# 6) Guardar palabras clave y su peso en un archivo de texto
output_path = f'CORFO_tendencias_{areas_oecd[choose]}.txt'
with open(output_path, 'w', encoding='utf-8') as f:
    f.write("palabra\tpeso\n")
    for palabra, peso in freq_dict.items():
        f.write(f"{palabra}\t{peso:.4f}\n")
print(f"Palabras clave guardadas en {output_path}")

# 7) Generar y mostrar la nube de palabras
wc = WordCloud(width=800, height=400, background_color='white')
wc.generate_from_frequencies(freq_dict)

plt.figure(figsize=(12, 6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title(
    f'Keywords: Área OECD {areas_oecd[choose]} ({num_projects} proyectos)',
    fontsize=16
)
plt.show()
```
